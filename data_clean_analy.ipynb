```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

df1 = pd.read_excel('Data_Train.xlsx')
df2 = pd.read_excel('Test_set.xlsx')
df = df1.append(df2)
df.head()
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python


df.drop_duplicates(keep="first",inplace=True)
```

```python
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'])
```

```python
df.info()
```

```python
df['Dep_Time'] = pd.to_datetime(df['Dep_Time'])
df['Arrival_Time'] = pd.to_datetime(df['Arrival_Time'])
```

```python
df.info()
```

```python
df['Journey_day'] = df['Date_of_Journey'].dt.day
df['Journey_month'] = df['Date_of_Journey'].dt.month
df['Journey_year'] = df['Date_of_Journey'].dt.year
df['Dep_hour'] = df['Dep_Time'].dt.hour
df['Dep_minute'] = df['Dep_Time'].dt.minute
df['Arrival_hour'] = df['Arrival_Time'].dt.hour
df['Arrival_minute'] = df['Arrival_Time'].dt.minute
```

```python
df.head()
```

```python
list_duration= list(df["Duration"])

for i in range(len(list_duration)):
    if len(list_duration[i].split()) !=2:
        if "h" in list_duration[i]:
            list_duration[i] = list_duration[i].strip() + " 0m"  
        else:
            list_duration[i] = "0h " + list_duration[i] 
            
hours = []
mins = []
for i in range(len(list_duration)):
    hours.append(int(list_duration[i].split(sep = "h")[0]))   
    mins.append(int(list_duration[i].split(sep = "m")[0].split()[-1])) 
df["Du_hours"] =hours
df["Du_mins"] =mins
```

```python
df.drop(['Date_of_Journey', 'Dep_Time', 'Arrival_Time', 'Duration', 'Route'], axis=1, inplace=True)
```

```python
df.dtypes
```

```python
df['Total_Stops']=df['Total_Stops'].replace('non-stop','0 stop')
```

```python
df['Total_Stops'] = df['Total_Stops'].str.split(' ').str[0]
df['Total_Stops']=df['Total_Stops'].astype(int)
```
